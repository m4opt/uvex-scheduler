```python
import ligo.skymap.plot  # noqa: F401
from cycler import cycler
from m4opt.fov._core import circle_to_polygon
from matplotlib import patches
from matplotlib import pyplot as plt
from regions import CircleSkyRegion, PointSkyRegion, Regions
```

```python
regions = [
    Regions.read("../survey-footprints/lmlz-wide.ds9"),
    Regions.read("../survey-footprints/lmlz-deep.ds9"),
    Regions.read("../survey-footprints/magellanic-clouds.ds9"),
]

fig = plt.figure()
ax = fig.add_subplot(projection="astro aitoff", center="8h 0d")
ax.set_facecolor("whitesmoke")
ax.grid()
handles = [patches.Patch(facecolor="whitesmoke", label="All-sky")]
for regions_, color_style, hatch_style, label in zip(
    regions,
    plt.rcParams["axes.prop_cycle"],
    cycler(hatch=r"/\|-+xo."),
    ["LMLZ Wide", "LMLZ Deep", "Magellanic"],
):
    in_legend = False
    for region in regions_:
        if isinstance(region, PointSkyRegion):
            ax.plot_coord(region.center, "+", **color_style)
        else:
            if isinstance(region, CircleSkyRegion):
                region = circle_to_polygon(region, 8)
            artist = ax.add_patch(
                region.to_pixel(ax.wcs).as_artist(
                    **color_style,
                    hatch=hatch_style["hatch"] * 3,
                    closed=True,
                    label=label,
                )
            )
            if not in_legend:
                handles.append(artist)
                in_legend = True
ax.legend(handles=handles)
fig.savefig("../visualizations/survey-footprints.pdf", metadata={"CreationDate": None})
```
